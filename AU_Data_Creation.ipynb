```python
#au_list represents list of AU's present in DISFA dataset
au_list = list([1,2,4,5,6,9,12,15,17,20,25,26])

#pair_list stores pair of frame numbers [offset, apex] for positive samples
#zero_list stores pair of frame numbers [offset, offset] for negative samples
#Each index of pair_list/zero_list correspond to AU number
pair_list = []
sub_no_list = []
for sub in glob.glob("**/*.txt", recursive=True):


  au_no = int(sub[32:-4])
  sub_no = int(sub[21:23])
  #print(au_no, sub_no)

  sub_no_list.append(sub_no)

  facs = np.loadtxt(sub, delimiter=',')
  test_list = list(facs[:, 1])
  freq = [(k, len(list(j))) for k, j in groupby(test_list)] #freq stores the consecutive element frequency of AU intensities
  freq = np.array(freq)
 
  index_zeros = [] #index_zeros saves the index of zeros in freq
  for i in range(freq.shape[0]):
   if freq[i,0]==0:
    index_zeros.append(i)
 
  
 
  for i in range(len(index_zeros)):
   
   #Handling the case when video doesn't start with 0 intensity frames
   if i==0:
    if index_zeros[i] != 0:
     offset = int(np.sum(freq[:(index_zeros[i]),1]))
     
     for index in range(0, index_zeros[i], 1):
    
      apex = np.sum(freq[:index, 1]) - 1 + ceil(freq[index,1]//2) 
      apex = int(apex)
      pair = [au_no, sub_no, offset, apex, freq[index,0]]
      pair_list.append(pair)

   #Handling the case when video doesn't end with 0 intensity frames
   if i==len(index_zeros)-1:
    if index_zeros[i] != freq.shape[0]-1:
     offset = int(np.sum(freq[:(index_zeros[i]+1),1])) - 1


     for index in range(index_zeros[i]+1, freq.shape[0] , 1):
    
      apex = np.sum(freq[:index, 1]) - 1 + ceil(freq[index,1]//2) 
      apex = int(apex)
      pair = [au_no, sub_no, offset, apex, freq[index,0]]
      pair_list.append(pair)
    continue
   
    #apex_index = np.argmax(facs[np.sum(freq[:,1][:index_zeros[i]]:np.sum(freq[:,1][:index_zeros(i+1)])[0]
   offset1 = int(np.sum(freq[:(index_zeros[i]+1),1])) - 1
   offset2 = int(np.sum(freq[:(index_zeros[i+1]),1]))

   
   for index in range(index_zeros[i]+1, index_zeros[i+1], 1):
    
    apex = np.sum(freq[:index, 1]) - 1 + ceil(freq[index,1]//2) 
    apex = int(apex)
    pair = [au_no, sub_no, offset1, apex, freq[index,0]]
    pair_list.append(pair)
    pair = [au_no, sub_no, offset2, apex, freq[index,0]]
    pair_list.append(pair)

print(pair_list)



   
 
```

### **Creating data with Image Registration/Normalization**

```python
df = pd.read_csv("AU_Data.csv", index_col=0)
```

```python
df.columns
```

```python
from skimage.transform import AffineTransform
x_n = np.zeros((df.shape[0], 272))

for i in range(df.shape[0]):
 
  z = np.array(df.iloc[i,5:277]).reshape(136,2)
  offset = z[:68, :]
  apex = z[68:, :]
  
  src = np.array([offset[0], offset[16], offset[27], offset[33], offset[39], offset[42]])
  dst = np.array([[-400,-400], [400, -400], [0, -400], [0, 0],[-100,-400],[100,-400]])
  tform1 = AffineTransform()
  tform1.estimate(src, dst)
  
  offset_n = tform1(offset)
  x_n[i, :136] = offset_n.flatten()

  src = np.array([apex[0], apex[16], apex[27], apex[33], apex[39], apex[42]])
  dst = np.array([[-400,-400], [400, -400], [0, -400], [0, 0],[-100,-400],[100,-400]])
  tform2 = AffineTransform()
  tform2.estimate(src, dst)
  
  apex_n = tform2(apex)
  x_n[i, 136:] = apex_n.flatten()
```

```python
column_values = []
au_list = list([1,2,4,5,6,9,12,15,17,20,25,26])


for i in range(68):
  column_values.append("Offset_Normalized_x"+str(i+1))
  column_values.append("Offset_Normalized_y"+str(i+1))

for i in range(68):
  column_values.append("Apex_Normalized_x"+str(i+1))
  column_values.append("Apex_Normalized_y"+str(i+1))


df3 = pd.DataFrame(data = x_n, columns = column_values)
```

```python
df4 = pd.concat([df, df3], axis=1)
```

```python
for i in range(12):
  temp = df4.pop('AU'+str(au_list[i]))
  df4['AU'+str(au_list[i])] = temp
```
